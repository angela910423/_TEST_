```python
content=np.random.randint(50,high=101,size=(50,5))
```

```python
import pandas as pd
pd.DataFrame(data=content, index=[f"第{i}號"for i in range(1,51)],columns=['國文','數學','英文','社會','自然'])
df1=pd.DataFrame(data=content, index=[f"第{i}號"for i in range(1,51)],columns=['國文','數學','英文','社會','自然'])
df1
```

```python
df1.loc["第1號"] #dataFrame.loc[key,欄] 使用key
```
